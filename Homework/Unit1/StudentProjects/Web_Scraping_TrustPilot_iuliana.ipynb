```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
all_companies =[]
all_num_rev_and_scores =[]
all_address =[]
all_cat = []
for i in range(1,6):
    my_req = requests.get(f'https://www.trustpilot.com/categories/shipping_logistics?page={i}')
    scrappy = BeautifulSoup(my_req.text)
    companies = scrappy.find_all('div', {'class': 'businessTitle___152-c'})
    num_rev_and_scores = scrappy.find_all('div', { 'class' : 'textRating___3F1NO'})
    address = scrappy.find_all('div', class_ = 'businessUnitCardBody___2SzlK')    
    cat = scrappy.find_all('div', {'class' :'categories___1-nja'})
    all_companies.extend(companies)
    all_num_rev_and_scores.extend(num_rev_and_scores)
    all_address.extend(address)
    all_cat.extend(cat)
    i +=1
```

```python
all_companies = [str(comp) for comp in all_companies]
all_companies = [comp.replace('</div>','') for comp in all_companies]
all_companies = [comp.split('>')[-1] for comp in all_companies]
```

```python
all_scores =[]
all_num_rev =[]
all_num_rev_and_scores = [str(rev) for rev in all_num_rev_and_scores]
all_num_rev_and_scores = [rev.replace('</div>','') for rev in all_num_rev_and_scores]
all_scores = [score.split()[-1] for score in all_num_rev_and_scores]
all_num_rev = [rev.split('>')[1] for rev in all_num_rev_and_scores]
all_num_rev = [rev.split()[0] for rev in all_num_rev]
```

```python
all_addresses =[]
all_address = [str(addr) for addr in all_address]
all_address = [addre.replace('</span></span></div></div>','') for addre in all_address]
for addr in all_address:
    if '<span class="locationZipcodeAndCity___33EfU">' in addr:
        all_addresses.append(addr.split('<span>')[-1])
    else:
        all_addresses.append('no address')      
```

```python
all_categ = []
all_cat = [str(cat) for cat in all_cat]
all_cat = [cat.replace('<div class="categories___1-nja"><span>','') for cat in all_cat]
all_cat = [cat.replace('</span> </div>','') for cat in all_cat]
for cat in all_cat:
    all_categ.append(cat.split('</span> <span class="categoriesBullet___1UNg0">·</span><span>'))
```

```python
dictionary1 = {
    'company': all_companies[0:100],
    '#reviews': all_num_rev[0:100],
    'rate': all_scores[0:100],
    'address': all_addresses[0:100],
    'category': all_categ[0:100]
}
df = pd.DataFrame(dictionary1)
df
```
